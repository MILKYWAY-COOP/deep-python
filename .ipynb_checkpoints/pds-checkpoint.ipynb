```python
#BASIC FUNCIONALITIES OF THE SERIES DATA STRUCTURE 
import pandas as pd
import numpy as np

#Create a series with 20 random numbers
s = pd.Series(np.random.randn(20))

print(s)
print('\n')

print("The axes are:")
print(s.axes)
print('\n')

print('Is the Object empty?')
print(s.empty)
print('\n')

print('The dimentions of the object:')
print(s.ndim)
print('\n')

print('The actual data series is:')
print(s.values)
print('\n')

print('The first four items of the data series:')
print(s.head(4))
print('\n')

print('The last four items of the data series:')
print(s.tail(5))
```

```python
# Adding rows to a Data Frame using the append function
import pandas as pd

df = pd.DataFrame([[1,2], [3,4]], columns = ['a', 'b'])
df2 = pd.DataFrame([[5,6], [7,8]], columns = ['a', 'b'], index = [2,3])

df = df.append(df2)
print(df)
print('\n')

# Drop rows with label 0
df = df.drop(0)
print(df)
```

```python
# Basic functionalities of the DATAFRAME data structure
import pandas as pd
import numpy as np

# Create a Dictionary of series
d = {'Name':pd.Series(['Tom', 'James', 'Ricky', 'Vin', 'Steve', 'Smith', 'Jack']),
    'Age':pd.Series([25, 26, 25, 23, 22, 29, 23]),
    'Rating':pd.Series([4.23, 3.24, 3.98, 2.56, 4.8, 4.6, 3.8])}

# Create a Data Frame
df = pd.DataFrame(d)
print('Our data Series is:')
print(df)
print('\n')

print('The transpose of the data series is:')
print(df.T)
print('\n')

print('Row axis labels and column axis labels are:')
print(df.axes)
print('\n')

print('The data types of each column are:')
print(df.dtypes)
print('\n')

print('Is the object empty?')
print(df.empty)
print('\n')

print('The dimesion of the object is:')
print(df.ndim)
print('\n')

print('The shape of the object is:')
print(df.shape)
print('\n')
      
print('The total number of elements in our object is:')
print(df.size)
print('\n')
      
print('The actual data in our data frame is:')
print(df.values)
print('\n')
      
print('The first two rows of the data frame is:')
print(df.head(2))
print('\n')
      
print('The last two rows of the data frame is:')
print(df.tail(2))
```

```python
# Sorting Data
import pandas as pd
import numpy as np

unsorted_df = pd.DataFrame(np.random.randn(10, 2), index=[1,4,6,2,3,5,9,8,0,7], columns=['col2', 'col1'])
print(unsorted_df)
print('\n')

#Sorting by index/label
sorted_df = unsorted_df.sort_index()
print(sorted_df)
print('\n')

#Sorting by values
print('Sorting by values')
unsorted_df = pd.DataFrame({'col1':[2,1,1,1],'col2':[1,3,2,4]})
print(unsorted_df)
sorted_df = unsorted_df.sort_values(by='col2')
print (sorted_df)
print('\n')

#Sorting using Mergesort
print('Sorting using Mergesort')
unsorted_df = pd.DataFrame({'col1':[2,1,1,1],'col2':[1,3,2,4]})
print(unsorted_df)
sorted_df = unsorted_df.sort_values(by='col1',kind='mergesort')
print (sorted_df)
```

```python
#Merging Data
import pandas as pd
left = pd.DataFrame({'id':[1,2,3,4,5],'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
                     'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame({'id':[1,2,3,4,5],'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
                      'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(left)
print(right)
print('\n')

#Merge the two data frames on a key
print('Merge the two data frames on a key')
print(pd.merge(left,right,on='id'))
print('\n')

#Merge the two data frames on multiple keys
print('Merge the two data frames on multiple keys')
print (pd.merge(left,right,on=['id','subject_id']))
print('\n')

#Left join
print('Left join')
print(pd.merge(left, right, on='subject_id', how='left'))
print('\n')

#Right join
print('Right Join')
print (pd.merge(left, right, on='subject_id', how='right'))
print('\n')

#Outer join
print('Outer Join')
print (pd.merge(left, right, how='outer', on='subject_id'))
print('\n')

#Inner join
print('Inner Join')
print (pd.merge(left, right, on='subject_id', how='inner'))
```
